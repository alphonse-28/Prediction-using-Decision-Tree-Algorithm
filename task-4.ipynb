# Step 2: Load the Iris Dataset & Exploring the data set

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()

# Convert the iris dataset to a pandas DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column to the DataFrame
iris_df['target'] = iris.target

# Display the first few rows of the DataFrame
print(iris_df.head())
```

# Visualization

```python


# Add the target column to the DataFrame with string labels
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Plotting Sepal Length vs Sepal Width
fig = iris_df[iris_df.species == 'setosa'].plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)', color='green', label='Setosa')
iris_df[iris_df.species == 'versicolor'].plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)', color='blue', label='Versicolor', ax=fig)
iris_df[iris_df.species == 'virginica'].plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)', color='orange', label='Virginica', ax=fig)

fig.set_xlabel("Sepal Length", fontsize=15)
fig.set_ylabel("Sepal Width", fontsize=15)
fig.set_title("Sepal Length Vs Sepal Width", fontsize=20)
plt.show()
```

```python
# Convert the iris dataset to a pandas DataFrame
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column to the DataFrame with string labels
data['Species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Renaming columns for convenience
data.columns = [col.replace(" (cm)", "Cm") for col in data.columns]

# Plotting Petal Length vs Petal Width
fig = data[data.Species == 'setosa'].plot.scatter(x='petal lengthCm', y='petal widthCm', color='green', label='Setosa')
data[data.Species == 'versicolor'].plot.scatter(x='petal lengthCm', y='petal widthCm', color='blue', label='Versicolor', ax=fig)
data[data.Species == 'virginica'].plot.scatter(x='petal lengthCm', y='petal widthCm', color='orange', label='Virginica', ax=fig)

fig.set_xlabel("Petal Length", fontsize=15)
fig.set_ylabel("Petal Width", fontsize=15)
fig.set_title("Petal Length Vs Petal Width", fontsize=20)
plt.show()
```

```python
# pairplot for better visualization
sns.pairplot(data, hue = 'Species')
```

# Step 3: Split the Dataset into Training and Testing Sets

```python
x=iris.data
y=iris.target
```

```python
print(x.shape)
print(y.shape)
```

# Step 4: Train the Decision Tree Model

```python
from sklearn import tree
df=tree.DecisionTreeClassifier()
df.fit(x,y)
```

# Step 5: Make Predictions

```python
# Make predictions on the testing data
y = df.predict(x)
```

# Step 6: Visualize the Decision Tree

```python
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(11,11),dpi=200)
_=tree.plot_tree(df,feature_names=iris.feature_names,class_names=iris.target_names,filled=True,fontsize=7)
```
